# The following Argivoltaic simulation was performed with Solar Edge and the assumptions are as follows  


*   Weather station: Walla Walla (at 31.39 km)
*   Elevation: 353 m

*   Number of modules: 73 331 
*   Number of inverters: 4 600 
*   Module model: Canadian Solar, CS3L-365MS
*   Azimuth: 21 degrees north
*   Tilt: 23 degrees



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Show the site and the modules placement
site = mpimg.imread('site.png')
imgplot = plt.imshow(site)
plt.show()

module = mpimg.imread('Modules.png')
imgplot = plt.imshow(module)
plt.show()
```

```python
# Create the power generation dataframe

df = pd.read_csv('data.csv')
```

```python
display(df)
```

```python
# Capacity and Generation
print('The capacity is 26.6 MWh and the total generation is ' + str(df['Generation (MWh)'].sum()) + ' MWh')
```

```python
# Create the generation Diagram 

ax = df.plot.bar(y='Generation (MWh)', x='Months', rot=0, color = 'g')
ax.set_xlabel('Months')
ax.set_ylabel('(MWh)')
plt.title('Plant generation over a year')    
plt.gcf().set_dpi(150)
plt.show()

```

```python
Factor = ['Tilt and Azimuth','Reflection','Temperature','Optimiser','DC', 'Inverter']
Loss= [21.17, 2.67, 2.28, 0.61, 0.59, 1.29]

plt.barh(Factor,Loss)
plt.title('Loss diagram')
plt.ylabel('Loss factor')
plt.xlabel('Loss (%)')

plt.gcf().set_dpi(150)

plt.show()
```

```python
# Now the Cash flow

df["Cash From Operation ($)"] = (0.12) * (1000)* df["Generation (MWh)"]
df["Lease Income ($)"] = (160) * (139)
df["Cash Flow ($)"] = df["Cash From Operation ($)"]+df["Lease Income ($)"]
display(df)
```

```python
# Cash Flow Diagram 

ax = df.plot.bar(y='Cash Flow ($)', x='Months', rot=0, color = 'orange')
ax.set_xlabel('Months')
ax.set_ylabel('($)')
plt.title('Cash Flow From Operation and Lease')    
plt.gcf().set_dpi(150)
```
